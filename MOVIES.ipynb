```python
import sqlite3

conn = sqlite3.connect('MOVIES.db')

print ("Opened database successfully");
```

```python
conn.execute('''CREATE TABLE Movies
         (MOVIE        TEXT    NOT NULL,
         ACTOR              TEXT    NOT NULL,
         ACTRESS            TEXT    NOT NULL,
         YEAR               REAL,
        DIRECTOR       TEXT);''')
print ("Table created successfully");
```

```python
cursor = conn.execute("SELECT MOVIE, ACTOR,ACTRESS,YEAR,DIRECTOR from Movies")
for row in cursor:
   print ("MOVIE = ", row[0])
   print ("ACTOR = ", row[1])
   print ("ACTRESS = ", row[2])
   print ("YEAR = ", row[3])
   print ("DIRECTOR = ", row[4]), "\n"

print ("Operation done successfully");
```
